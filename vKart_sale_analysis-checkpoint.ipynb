```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = sns.countplot(x = 'Gender', data= df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y = 'Amount' , data= sales_gender)
```

```python
 bx = sns.countplot(x='Age Group' , hue= "Gender" , data = df)
    
for bars in bx.containers:
        bx.bar_label(bars)
```

```python
 #From this graph we can see that most of the buyers are female age between 26-35 yrs
```

```python
sales_age_group = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending = False)
sns.barplot(x="Age Group", y = 'Amount', data=sales_age_group)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders' , ascending = False)
sns.set(rc={'figure.figsize' : (25,5)})
sns.barplot(x='State', y = "Orders", data= sales_state)
```

```python
# From this graph we can see that Uttar Pradesh has most number of orders
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending = False)
sns.set(rc={'figure.figsize' : (25,5)})
sns.barplot(x='State', y = "Amount", data= sales_state)
```

```python
# From this graph we can see that which state has the highest amount 
```

```python
cx =sns.countplot(data=df, x='Marital_Status' , hue='Gender')
sns.set(rc={'figure.figsize' : (10,5)})
for bars in cx.containers:
    cx.bar_label(bars)
```

```python
dx = sns.countplot(x='Occupation', data =df)
sns.set(rc={'figure.figsize' : (30,5)})
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
sales_by_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending = False)
sns.set(rc={'figure.figsize' : (25,5)})
sns.barplot(x='Occupation', y = "Amount", data= sales_by_occupation)
```

```python
#From this graph we can see that most amount spent by IT sector , Healthcare , Aviation
```

```python
sales_productctgr = df.groupby(['Product_Category'],  as_index=False)['Amount'].sum().sort_values(by = 'Amount' , ascending=False).head(10)
sns.set(rc={'figure.figsize' : (30,10)})
sns.barplot(data = sales_productctgr, x = 'Product_Category' , y = 'Amount')
```

```python
#From above graphs we can see that most of the sold product are from Food , Clothing & Apparel and Electronics & Gadgets 
```
